```python
import flux.job
from pyiron_atomistics import Project
```

```python
pr = Project('flux')
pr.remove_jobs(recursive=True, silently=True)
```

```python
structure = pr.create.structure.ase.bulk("Al", cubic=True)
```

# Use flux as queuing system

```python
job = pr.create.job.Lammps("lmp_with_queue")
job.executable.version = "2020.03.03_flux"
job.structure = structure
job.server.queue = "flux"
job.server.cores = 2
job.run()
```

```python
pr.wait_for_job(job)
```

# Use flux Executor

```python
with flux.job.FluxExecutor() as exe:
    job = pr.create.job.Lammps("lmp_with_exe")
    job.executable.version = "2020.03.03_flux"
    job.structure = structure
    job.server.executor = exe
    job.server.cores = 2
    job.run()
    print(job.server.future.result())
```
